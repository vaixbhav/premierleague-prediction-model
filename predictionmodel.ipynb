```python
matches["date"] = pd.to_datetime(matches["date"])       #Converting date's object datatype to datetime data 
```

```python
matches["target"] = (matches["result"] == "W").astype("int")        #Convert match result from W(win) to 1, L(loss) to 0, D(draw) to 0
```

```python
matches
```

```python
matches["venue_code"] = matches["venue"].astype("category").cat.codes   #Convert venue_code string datatype to category and then to numerical values
```

```python
matches["opp_code"] = matches["opponent"].astype("category").cat.codes  #Creating code for each team
```

```python
matches["hour"] = matches["time"].str.replace(":.+", "", regex=True).astype("int")      #Replace hour datatype to integer datatype
```

```python
matches["day_code"] = matches["date"].dt.dayofweek      #Code for each day of the week
```

```python
matches
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
#Random Forest contains 50 decision trees and will not split internal nodes under 10 trees
rf = RandomForestClassifier(n_estimators=50, min_samples_split=10, random_state=1)  
```

```python
train = matches[matches["date"] < '2023-11-01']     #Dataset used to train machine learning model
```

```python
test = matches[matches["date"] > '2023-11-01']      #Test dataset to evaluate precision of trained dataset 
```

```python
predictors = ["venue_code", "opp_code", "hour", "day_code"]     #List of predictor columns
```

```python
rf.fit(train[predictors], train["target"])      #Train random forest model with predictors to predict target (win: 1, loss/draw: 0)
```

```python
preds = rf.predict(test[predictors])    
```

```python
from sklearn.metrics import accuracy_score
```

```python
acc = accuracy_score(test["target"], preds)     #Determine accuracy of results predicted
```

```python
acc
```

```python
combined = pd.DataFrame(dict(actual=test["target"], predicted=preds))       #Analyze whether wins/losses were predicted more accurately
```

```python
pd.crosstab(index=combined["actual"], columns=combined["predicted"])   #Predicted 55/91 losses and 32/64 wins
```

```python
from sklearn.metrics import precision_score     #Precision score (percentage of wins)

precision_score(test["target"], preds)
```

```python
grouped_matches = matches.groupby("team")       #Create DataFrame for every squad in dataset 
```

```python
group = grouped_matches.get_group("Liverpool").sort_values("date")  #Analyze Liverpool's data
```

```python
def rolling_averages(group, cols, new_cols):        #Compute rolling average based on previous 3 match results
    group = group.sort_values("date")
    rolling_stats = group[cols].rolling(3, closed='left').mean()
    group[new_cols] = rolling_stats
    group = group.dropna(subset=new_cols)           #In the case of 3 match result data not available
    return group
```

```python
cols = ["gf", "ga", "sh", "sot", "dist", "fk", "pk", "pkatt"]       #Columns to compute rolling averages
new_cols = [f"{c}_rolling" for c in cols]

rolling_averages(group, cols, new_cols)
```

```python
matches_rolling = matches.groupby("team").apply(lambda x: rolling_averages(x, cols, new_cols))  #Apply rolling averages for all matches
```

```python
matches_rolling
```

```python
matches_rolling = matches_rolling.droplevel('team')
```

```python
matches_rolling
```

```python
matches_rolling.index = range(matches_rolling.shape[0])     #Ensuring unique values for each index
```

```python
def make_predictions(data, predictors):         #Iterate prediction model with date boundaries and predictors   
    train = data[data["date"] < '2023-11-01']
    test = data[data["date"] > '2023-11-01']
    rf.fit(train[predictors], train["target"])
    preds = rf.predict(test[predictors])
    combined = pd.DataFrame(dict(actual=test["target"], predicted=preds), index=test.index)
    error = precision_score(test["target"], preds)
    return combined, error
```

```python
combined, error = make_predictions(matches_rolling, predictors + new_cols)      #Check error 
```

```python
error
```

```python
#Merge DataFrames to show stats based on index 
combined = combined.merge(matches_rolling[["date", "team", "opponent", "result"]], left_index=True, right_index=True)
```

```python
class MissingDict(dict):                    #Dictionary to ensure if team name missing, it is appropriately replaced
    __missing__ = lambda self, key: key

map_values = {"Brighton and Hove Albion": "Brighton", "Manchester United": "Manchester Utd", "Newcastle United": "Newcastle Utd", "Tottenham Hotspur": "Tottenham", "West Ham United": "West Ham", "Wolverhampton Wanderers": "Wolves"} 
mapping = MissingDict(**map_values)
```

```python
combined["new_team"] = combined["team"].map(mapping)        #Same team names on home and opponent columns
```

```python
#Merge dataframes for same match (ex. Arsenal vs. Chelsea match is same as Chelsea vs. Arsenal match)
merged = combined.merge(combined, left_on=["date", "new_team"], right_on=["date", "opponent"])
```
